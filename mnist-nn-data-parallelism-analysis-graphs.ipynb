```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt


import os

if os.path.exists('/tmp/outputs') is False:
    os.mkdir('/tmp/outputs')
```

```python
N = 4
ind = np.arange(N)  # the x locations for the groups
width = 0.25       # the width of the bars

cpu_local_times = (0, 0.05, 0.43, 0.55)
cpu_dist_times = (0.07, 0.03, 0.22, 0.38)
gpu_local_times = (0, 0.11, 1.61, 1.75)
gpu_dist_times = (0.07, 0.05, 0.81, 1.00)

xlabels = ('scatter', 'Forward Propagation', 'Back Propagation', 'Total')

for i in xrange(1, N):
    fig, ax = plt.subplots()

    ind = 0.0
    scalar_base = min(cpu_local_times[i], cpu_dist_times[i], gpu_local_times[i], gpu_dist_times[i])
    rects1 = ax.bar(ind + width*0, cpu_local_times[i]/scalar_base, width, color='indianred')
    rects2 = ax.bar(ind + width*1, cpu_dist_times[i]/scalar_base,  width, color='royalblue')
    rects3 = ax.bar(ind + width*2, gpu_local_times[i]/scalar_base, width, color='slategrey')
    rects4 = ax.bar(ind + width*3, gpu_dist_times[i]/scalar_base,  width, color='darkcyan')

    # add some text for labels, title and axes ticks
    #ax.set_xlabel(xlabels[i])
    ax.set_xticks([ind + width*2])
    ax.set_xticklabels([xlabels[i]], fontsize=14)
    ax.set_ylabel('Execution Time', fontsize=14)
    ax.set_title('Data parallelism performance comparison', fontsize=16)
    ax.legend(
        (rects1[0], rects2[0], rects3[0], rects4[0]),
        ('Single CPU', 'Parallelism on 2 CPUs', 'Single GPU', 'Parallelism on 2 GPUs'),
        loc='upper left'
        #bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.
        #bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.
    )


    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(
                rect.get_x() + rect.get_width()/2.,
                #1.05*height,
                1.0*height,
                '%.2f' % height,
                ha='center', va='bottom', fontsize=14
            )

            
    autolabel(rects1)
    autolabel(rects2)
    autolabel(rects3)
    autolabel(rects4)

    param = plt.gcf()
    pltsize = param.get_size_inches()
    param.set_size_inches(pltsize[0]*1.5, pltsize[1]*1.5)
    param.savefig('/tmp/outputs/' + xlabels[i].lower().replace(' ', '-'))
    plt.show()
```
